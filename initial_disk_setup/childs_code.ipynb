```python
#Some constants
# Define constants
# Define constants

sol_to_earth = 332946.078
AU = 1.496e+13 #cm
M_sun = 1.989e+33 #g

a_min = 0.35  # Minimum semi-major axis 
a_max = 4.0   # Maximum semi-major axis

n_pl = 260    # Number of planetesimals
n_emb = 26    # Number of planetary embryos

m_pl = 2.7932451e-8  # Mass of each planetesimal
m_emb = 2.7932451e-7  # Mass of each planetary embryo

r_emb = 2.39e-5 #radius of embryo
r_pl = 1.11e-5 #radius of planetesimal

n_steps = 100 #steps of dividing the disk

m_tot = n_pl * m_pl + n_emb * m_emb #total mass in the disk
```

```python
def cum_mass(a, body_mass, body_n, n_steps, a_min, a_max):
    x = np.linspace(a_min, a_max, n_steps)
    delta_m = np.zeros(n_steps)
    delta_sigma = np.zeros(n_steps)
    m_sum = np.zeros(n_steps)
    
    # Calculate areas of each annulus
    areas = np.pi * (x[1:]**2 - x[:-1]**2)

    # Assign body mass to each annulus
    indices = np.digitize(a, x) - 1  # Get index of annulus for each body
    np.add.at(delta_m, indices, body_mass)
    
    m_sum = np.cumsum(delta_m)
    
    return m_sum
```
